```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

### Experiment 1

Configurations:
- Dataset size: 1 sample
- Model
    - Improved Visconet
    - Latent: Target Image,
    - Styles: Target Image,
    - CLIP
    - Number of queries: 16
    - Number of fashion attributes: 5 

```python
expt1_log_parent_dir = os.path.join(os.getcwd(), "..", "logs", "test-visconet-improved-target-latents-target-style-clip-numqueries-16-numattrs-5-oneimg", "lightning_logs", "version_47283", "metrics_logs")

expt1_train_loss_history = os.path.join(expt1_log_parent_dir, "train_loss_epoch.csv")
expt1_train_loss_history_df = pd.read_csv(expt1_train_loss_history)

expt1_val_loss_history = os.path.join(expt1_log_parent_dir, "val_loss.csv")
expt1_val_loss_history_df = pd.read_csv(expt1_val_loss_history)

fig, axs = plt.subplots(1, 3, figsize=(16,8))

axs[0].plot(expt1_train_loss_history_df["Step"], expt1_train_loss_history_df["Value"])
axs[0].set_title("Train Loss History")
axs[1].plot(expt1_val_loss_history_df["Step"], expt1_val_loss_history_df["Value"])
axs[1].set_title("Validation Loss History")

axs[1].set_title("Combined Loss History")
axs[2].plot(expt1_train_loss_history_df["Step"], expt1_train_loss_history_df["Value"], label="train")
axs[2].plot(expt1_val_loss_history_df["Step"], expt1_val_loss_history_df["Value"], label="val")
axs[2].legend()
```

### Experiment 2

Configurations:
- Dataset size: Full Dataset (2,000 samples)
- Model
    - Baseline Visconet
    - Latent: Target Image,
    - Styles: Target Image

```python
expt2_log_parent_dir_run1 = os.path.join(os.getcwd(), "..", "logs", "test-visconet-baseline-target-latent-target-style-fullds", "lightning_logs", "version_85514", "metrics_logs")
expt2_log_parent_dir_run2 = os.path.join(os.getcwd(), "..", "logs", "test-visconet-baseline-target-latent-target-style-fullds-resumept2", "lightning_logs", "version_93851", "metrics_logs")

expt2_train_loss_run1 = os.path.join(expt2_log_parent_dir_run1, "train_loss_epoch.csv")
expt2_train_loss_run1_df = pd.read_csv(expt2_train_loss_run1)
expt2_val_loss_run1 = os.path.join(expt2_log_parent_dir_run1, "val_loss_epoch.csv")
expt2_val_loss_run1_df = pd.read_csv(expt2_val_loss_run1)

expt2_train_loss_run2 = os.path.join(expt2_log_parent_dir_run2, "train_loss_epoch.csv")
expt2_train_loss_run2_df = pd.read_csv(expt2_train_loss_run2)
expt2_train_loss_run2_df["Step"] += max(expt2_train_loss_run1_df["Step"])
expt2_train_loss_df = pd.concat([expt2_train_loss_run1_df, expt2_train_loss_run2_df], axis=0)

expt2_val_loss_run2 = os.path.join(expt2_log_parent_dir_run2, "val_loss_epoch.csv")
expt2_val_loss_run2_df = pd.read_csv(expt2_val_loss_run2)
expt2_val_loss_run2_df["Step"] += (max(expt2_val_loss_run1_df["Step"]))
expt2_val_loss_df = pd.concat([expt2_val_loss_run1_df, expt2_val_loss_run2_df], axis=0)

fig, axs = plt.subplots(1, 3, figsize=(16,8))

axs[0].plot(expt2_train_loss_df["Step"], expt2_train_loss_df["Value"])
axs[0].set_title("Train Loss History")
axs[1].plot(expt2_val_loss_df["Step"], expt2_val_loss_df["Value"])
axs[1].set_title("Validation Loss History")

axs[2].set_title("Combined Loss History")
axs[2].plot(expt2_train_loss_df["Step"], expt2_train_loss_df["Value"], label="train")
axs[2].plot(expt2_val_loss_df["Step"], expt2_val_loss_df["Value"], label="val")
axs[2].legend()
```
